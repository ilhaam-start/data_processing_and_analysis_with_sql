### Challenge

Write a query which returns the difference between the "High Price" and the "Low Price" of pumpkins in different cities. Ensure that the results are returned in descending order.

When your query is written correctly, you should see the following results:

|City Name      |	Difference      |
|--------------:|-------------------|
|BOSTON	        |24.394886363636374 |
|ATLANTA	    |14.232456140350877 |
|NEW YORK	    |9.357142857142861  |
|BALTIMORE   	|7.56535947712419   |
|LOS ANGELES    |5.812096774193563  |
|PHILADELPHIA	|5.614035087719287  |
|SAN FRANCISCO	|5.25               |
|DALLAS	        |4.81751824817519   |
|CHICAGO	    |2.665322580645153  |
|DETROIT        |1.2272727272727195 |
|ST. LOUIS	    |1.1407766990291321 |
|COLUMBIA	    |0.41825095057035355|
|MIAMI	        |0                  |


```python
%%sql

SELECT "City Name", (SUM("High Price") - SUM("Low Price")) / COUNT("City Name") AS "Difference" 
FROM us_pumpkin_prices
GROUP BY "City Name"
ORDER BY "Difference" DESC
```

## JOINS

```python
%%sql

SELECT * 
FROM athletes 
INNER JOIN medals 
ON athletes.name = medals.name
LIMIT 5
```

```python
%%sql

SELECT athletes.name, athletes.date_of_birth, medals.medal
FROM athletes INNER JOIN medals
ON athletes.name = medals.name AND athletes.sport = medals.sport
ORDER BY athletes.date_of_birth
LIMIT 5;
```

```python
%%sql

SELECT athletes.name, athletes.date_of_birth, medals.medal
    FROM athletes INNER JOIN medals
    ON athletes.name = medals.name AND athletes.sport = medals.sport
    ORDER BY athletes.date_of_birth DESC
    LIMIT 5;
```

```python
%%sql

SELECT a1.name, a2.name, a1.date_of_birth
FROM athletes_short AS a1
JOIN athletes_short AS a2 ON a1.date_of_birth = a2.date_of_birth
```
